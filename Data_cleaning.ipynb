### Import Libraries

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/data_cleaning.csv')
df
```

for instance if he didn't pick up correctly

```python
missing_value = ['Nan','nan', np.nan]

df = pd.read_csv('/content/data_cleaning.csv', na_values=missing_value)
```

```python
df
```

This is another method to pick up all the NaN values

### Remove this NaN values

```python
df.dropna()
```

This way we can remove all the nan values

```python
df.dropna(how = 'all')
```

How = all, works by removing all those values which have all NaN in a row

That's why it didn't remove any

```python
df.dropna(how='any')
```

how = any, works by removing any NaN values in the dataset

### we can fill this NaN data as well

```python
df.fillna(0)
```

In this way all the NaN value is filled with 0

```python
df.fillna(method = 'ffill')
```

ffill = forward fill, it takes value from the forward row

```python
df.fillna(method = 'bfill')
```

Bfill = backward fill

There is one method interpolate, through that we can get the missing value , backward + forward / 2

```python
df.interpolate()
```
